### Top 4 country have most population

```python
Top4pop = pd.DataFrame(pop[0:4])
# sum of pop
# for i in range(len(Top4pop)):
#   sum += Top4pop.loc[i, 'Population (2020)']
# set size in inches
fig, ax = plt.subplots(figsize=(16, 5))
#circle
colors = ['#306dbf', '#4884d4', '#6ea3eb', '#b0d2ff']
    # create pie
plt.pie(Top4pop['Population (2020)'], startangle=90, autopct='%1.1f%%', 
        counterclock=False, colors=colors, 
        wedgeprops={'edgecolor': 'w', 'linewidth': 1},)
    # Note for pie
plt.legend(Top4pop['Country (or dependency)'], loc='center right', bbox_to_anchor=(1.35, 0.7))

# title
plt.suptitle(r"$\bf{" + "TOP 4" + "}$" + ' country have most population', fontsize=24, color='#444444', x=0.35, ha='left')
# plt.title(f'Sum of population: {sum} ', fontsize=18, color='#777777', y=0.9, ha='center')
plt.savefig('pop.png')
```

### Top 4 country have most net Change

```python
pop1 = pop.sort_values(by='Net Change', ascending=False)
net = pop1[0:4]

fig, ax = plt.subplots(figsize=(16, 5))
#circle
colors = ['#306dbf', '#4884d4', '#6ea3eb', '#b0d2ff']
    # create pie
plt.pie(net['Net Change'], startangle=90, autopct='%1.1f%%', 
        counterclock=False, colors=colors, 
        wedgeprops={'edgecolor': 'w', 'linewidth': 1},)
    # Note for pie
plt.legend(net['Country (or dependency)'], loc='center right', bbox_to_anchor=(1.35, 0.7))

# title
plt.suptitle(r"$\bf{" + "TOP 4" + "}$" + ' country have most net Change', fontsize=24, color='#444444', x=0.35, ha='left')
# plt.title(f'Sum of population: {sum} ', fontsize=18, color='#777777', y=0.9, ha='center')
plt.savefig('pop.png')

net
```

### MDDS of 4 country

```python
Top4den = pd.DataFrame(pop[0:4])
x = Top4den['Country (or dependency)']
y = Top4den['Density (P/Km²)']

# set bar
fig, ax = plt.subplots(figsize=(10, 6))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


#create bar
plt.bar(x, y, width=0.6)
# box on bar
for i in range(len(Top4den)):
    plt.text(i, y[i], y[i], ha = 'center', fontsize=12,
              Bbox = dict(facecolor = 'yellow', alpha =.8)) # alpha: độ mờ

# title
plt.suptitle('Show the different density of 4 country', fontsize=24, color='#444444', x=0.1, ha='left')
plt.title('Density (P/Km²)', fontsize=18, color='#777777', y=0.9,x=0.85)
plt.xlabel('Country', fontsize=18, color='#333333')
plt.ylabel('Density Change ', fontsize=18, color='#333333')
# plt.savefig('MDDS.png')
```

### Top 4 country have largest density 

```python
pop2 = pop.sort_values(by='Density (P/Km²)', ascending=False)
den = pd.DataFrame(pop2[0:4])
x = den['Country (or dependency)']
y = den['Density (P/Km²)']

# set bar
fig, ax = plt.subplots(figsize=(10, 6))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


#create bar
plt.bar(x, y, width=0.6)

# title
plt.suptitle('Top 4 country have largest density', fontsize=24, color='#444444', x=0.1, ha='left')
plt.title('Density (P/Km²)', fontsize=18, color='#777777', y=0.9,x=0.85)
plt.xlabel('Country', fontsize=18, color='#333333')
plt.ylabel('Density Change (People)', fontsize=18, color='#333333')
```
