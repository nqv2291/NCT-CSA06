```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
population = pd.read_csv('population.csv').fillna('Not Given').replace('N.A.', 'Not Given')
population.insert(0, 'Index', [i+1 for i in range(population.shape[0])])
display(population)
```

```python
# Population
fig, ax = plt.subplots(figsize=(16, 6))

plt.bar(population['Country (or dependency)'][:20],
        population['Population (2020)'][:20], width=0.6, tick_label=population['Index'][:20])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.margins(x=0)
plt.suptitle('Population Per Country',
             fontsize=24, color='#444444')
plt.title('The Most Populated 20', fontsize=18, color='#777777')
plt.xlabel('Country', fontsize=18, color='#333333')
plt.ylabel('Population (People)', fontsize=18, color='#333333')

plt.show()
```

```python
# Population
fig, ax = plt.subplots(figsize=(16, 6))

plt.bar(population['Country (or dependency)'][-20:],
        population['Population (2020)'][-20:], width=0.6, tick_label=population['Index'][-20:])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.margins(x=0)
plt.suptitle('Population Per Country',
             fontsize=24, color='#444444')
plt.title('The Least Ppopulated 20', fontsize=18, color='#777777')
plt.xlabel('Country', fontsize=18, color='#333333')
plt.ylabel('Population (People)', fontsize=18, color='#333333')

plt.show()
```

```python
# Population Change

def mod_percentage(p):
  p = str(p)
  p = p[:-1]
  return eval(p)

pop_rate_change = population.copy()
pop_rate_change['Yearly Change'] = [mod_percentage(
    i) for i in pop_rate_change['Yearly Change']]
pop_rate_change.sort_values(by=['Yearly Change'], inplace=True, ascending=False)
rate_plot = pd.concat([pop_rate_change[:20], pop_rate_change[-20:]], ignore_index=True)

fig, ax = plt.subplots(figsize=(16, 6))

plt.bar(rate_plot['Country (or dependency)'],
        rate_plot['Yearly Change'], width=0.6, color="#009900", tick_label=rate_plot['Index'])
plt.bar(rate_plot['Country (or dependency)'][20:],
        rate_plot['Yearly Change'][20:], width=0.6, color="red")

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.margins(x=0)
plt.suptitle('Population Percentage Change Per Country',
             fontsize=24, color='#444444')
plt.title('The Highest and Lowest 20', fontsize=18, color='#777777')
plt.xlabel('Country', fontsize=18, color='#333333')
plt.ylabel('Population Change (%)', fontsize=18, color='#333333')

plt.show()
```

```python
# Population Change

pop_net_change = population.copy()
pop_net_change.sort_values(
    by=['Net Change'], inplace=True, ascending=False)
net_plot = pd.concat(
    [pop_net_change[:20], pop_net_change[-20:]], ignore_index=True)

fig, ax = plt.subplots(figsize=(16, 6))

plt.bar(net_plot['Country (or dependency)'][:20],
        net_plot['Net Change'][:20], width=0.6, color="#009900", tick_label=net_plot['Index'][:20])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.margins(x=0)
plt.suptitle('Population Net Change Per Country',
             fontsize=24, color='#444444')
plt.title('The Highest 20', fontsize=18, color='#777777')
plt.xlabel('Country', fontsize=18, color='#333333')
plt.ylabel('Population Change (People)', fontsize=18, color='#333333')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 6))

plt.bar(net_plot['Country (or dependency)'][-20:],
        net_plot['Net Change'][-20:], width=0.6, color="red", tick_label=net_plot['Index'][-20:])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.margins(x=0)
plt.suptitle('Population Net Change Per Country',
             fontsize=24, color='#444444')
plt.title('The Lowest 20', fontsize=18, color='#777777')
plt.xlabel('Country', fontsize=18, color='#333333')
plt.ylabel('Population Change (People)', fontsize=18, color='#333333')

plt.show()
```

```python
# Population Density

pop_dense = population.copy()
pop_dense.sort_values(
    by=['Density (P/Km²)'], inplace=True, ascending=False)

fig, ax = plt.subplots(figsize=(16, 6))

plt.bar(pop_dense['Country (or dependency)'][:20],
        pop_dense['Density (P/Km²)'][:20], width=0.6, tick_label=pop_dense['Index'][:20])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.margins(x=0)
plt.suptitle('Population Density Per Country',
             fontsize=24, color='#444444')
plt.title('The Most Dense 20', fontsize=18, color='#777777')
plt.xlabel('Country', fontsize=18, color='#333333')
plt.ylabel('Population Density (P/Km²)', fontsize=18, color='#333333')

plt.show()
```

```python
# Migrants

migrants = population.copy()
for i, row in migrants.iterrows():
  if row["Migrants (net)"] == 'Not Given':
    migrants.drop(i, inplace=True)
migrants.sort_values(by=["Migrants (net)"], inplace=True, ascending=False)
migrants_plot = pd.concat([migrants[:20], migrants[-20:]], ignore_index=True)

fig, ax = plt.subplots(figsize=(16, 6))

plt.bar(migrants_plot['Country (or dependency)'],
        migrants_plot['Migrants (net)'], width=0.6, color="#009900", tick_label=migrants_plot['Index'])
plt.bar(migrants_plot['Country (or dependency)'][-20:],
        migrants_plot['Migrants (net)'][-20:], width=0.6, color="red")

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.margins(x=0)
plt.suptitle('Migrants Per Country',
             fontsize=24, color='#444444')
plt.title('The Highest and Lowest 20', fontsize=18, color='#777777')
plt.xlabel('Country', fontsize=18, color='#333333')
plt.ylabel('Migrant (People)', fontsize=18, color='#333333')

plt.show()
```

```python
# Median age

median = population.copy()
for i, row in median.iterrows():
  if row["Med. Age"] == 'Not Given':
    median.drop(i, inplace=True)
median.sort_values(by=["Med. Age"], inplace=True, ascending=False)
median_plot = pd.concat([median[:20], median[-20:]], ignore_index=True)

fig, ax = plt.subplots(figsize=(16, 6))

plt.bar(median_plot['Country (or dependency)'][:20],
        median_plot['Med. Age'][:20], width=0.6, color="#ff6600", tick_label=median_plot['Index'][:20])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.margins(x=0)
plt.suptitle('Country\'s Median Age',
             fontsize=24, color='#444444')
plt.title('The Highest 20', fontsize=18, color='#777777')
plt.xlabel('Country', fontsize=18, color='#333333')
plt.ylabel('Rounded Median Age', fontsize=18, color='#333333')

plt.show()

```
