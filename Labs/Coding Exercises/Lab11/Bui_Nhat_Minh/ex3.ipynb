```python
import pandas as pd

border_dict = pd.read_excel('borders.xls', sheet_name=None)
border_chn = pd.DataFrame(border_dict['CHN'])
border_lao = pd.DataFrame(border_dict['LAO'])
border_khm = pd.DataFrame(border_dict['KHM'])

border_provinces = {}
border_count = {}

def add_bordering_countries(df, country):
  for index, row in df.iterrows():
    province = row['Tên tỉnh'] 
    if province in border_provinces:
      border_provinces[province] += f', {country}'
      border_count[province] += 1
    else:
      border_provinces[province] = country
      border_count[province] = 1
add_bordering_countries(border_chn, 'China')
add_bordering_countries(border_lao, 'Laos')
add_bordering_countries(border_khm, 'Cambodia')

df = pd.concat([pd.Series(border_provinces), pd.Series(border_count)], axis=1).reset_index().set_axis(['Name', 'BorderWith', 'BorderCount'], axis=1)
display(df)

df.to_excel('./borders_updated.xlsx')
```
