```python
import pandas as pd
df = pd.read_excel('provinces.xls')
df['PopulationDensity'] = df['Population'] / df['Area'] * 1000
cl = pd.read_excel('coastlines.xls')
print(f'Number of Coastal Provinces: {cl.shape[0]}')

df = pd.merge(df, cl, how='left').fillna(0)
display(df)

df.to_excel('./provinces_updated.xlsx')
```
