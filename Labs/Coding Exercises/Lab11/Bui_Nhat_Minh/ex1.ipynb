```python
import pandas as pd
df = pd.read_excel('provinces.xls')
df['PopulationDensity'] = df['Population'] / df['Area'] * 1000
display(df)
```

```python
max_population_density = df.loc[df['PopulationDensity'] ==
                                df['PopulationDensity'].max(), ['Name', 'PopulationDensity']]
display(max_population_density)
min_population_density = df.loc[df['PopulationDensity'] ==
                                df['PopulationDensity'].min(), ['Name', 'PopulationDensity']]
display(min_population_density)
```
