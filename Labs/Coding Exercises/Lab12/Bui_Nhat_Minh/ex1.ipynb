```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('./provinces_updated.xlsx').sort_values('Population').reset_index().iloc[:, 2:]
```

```python
fig, ax = plt.subplots(figsize=(16, 6))
city_index = [i+1 for i in range(df.shape[0])]
plt.bar(df['Name'], df['Population'], tick_label=city_index)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.margins(0.01, 0)
plt.suptitle('Populations in Vietnam Provinces and Cities',
             fontsize=24, color='#444444', x=0.125, ha='left')
plt.title('sorted ascending', fontsize=18, color='#777777', x=0, ha='left')
plt.xlabel('Province / City', fontsize=18, color='#333333')
plt.ylabel('Population (1000 people)', fontsize=18, color='#333333')
plt.plot()
print('Index of provinces and cities:', end=' ')
for i in range(df.shape[0]):
  city = df['Name'][i]
  print(f'{i+1}. {city}', end=' ')
```
