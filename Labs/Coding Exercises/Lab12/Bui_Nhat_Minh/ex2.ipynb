```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('./region_areas.xls')
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(df['Area Proportion (%)'], labels=df['Region'], autopct='%1.1f%%',
        startangle=90, counterclock=False,  wedgeprops={'edgecolor': 'w', 'linewidth': 1})
ax.set_title('Proportion of Area in Regions', fontsize=18, color='#333333')
plt.plot()
```
