```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('./provinces_updated.xlsx')
```

```python
fig, ax = plt.subplots(figsize=(16, 6))

n, bins, edges = plt.hist(df['PopulationDensity'], bins=20, ec='white')
plt.xticks(bins)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.suptitle('Frequency of Population Density in Vietnam Provinces and Cities',
             fontsize=24, color='#444444')
plt.xlabel('Population Density (people/km2)', fontsize=18, color='#333333')
plt.ylabel('Frequency', fontsize=18, color='#333333')
plt.show()
```
