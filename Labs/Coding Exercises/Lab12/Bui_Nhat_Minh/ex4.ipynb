```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./cyclones.csv')
display(df)
```

```python
fig, ax = plt.subplots(figsize=(16, 6))

plt.plot(df['date'], df['cyclone_events'])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_title('Cyclone Events in Vietnam',
          fontsize=24, color='#444444', x=0.125, ha='left')
ax.set_xlabel('Year', fontsize=18, color='#333333')
ax.set_ylabel('Number of Events', fontsize=18, color='#333333')

plt.plot()
```
